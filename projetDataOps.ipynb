```python
# Import des bibliothèques
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les données depuis le fichier CSV
df = pd.read_csv("collecte.csv")
```

```python
# Afficher les premières lignes du dataframe pour vérifier le chargement des données
print(df.head())
```

```python
# Résumé statistique des données
print(df.describe())
```

```python
# Compter le nombre de phrases par catégorie
sentences_count = df['sentences'].value_counts()
```

```python
# Tracer un diagramme à barres pour visualiser les comptages de phrases
plt.figure(figsize=(10, 6))
sentences_count.plot(kind='bar')
plt.title('Nombre de phrases par catégorie')
plt.xlabel('Catégorie')
plt.ylabel('Nombre de phrases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Tracer un nuage de points de la date par rapport au nombre de phrases
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='date', y='sentences')
plt.title('Nuage de points de la date par rapport au nombre de phrases')
plt.xlabel('Date')
plt.ylabel('Phrases')
plt.xticks(rotation=45)
plt.show()
```
